```python
import numpy as np
from scipy import signal
import math
```

```python
value_map = {
	0 : [1,0,0,0],
	1 : [0,1,0,0],
	2 : [0,0,1,0],
	3 : [0,0,0,1]
}

func = np.vectorize(lambda x:math.floor(x*2))
```

```python
a = func(np.random.rand(20))
b = func(np.random.rand(20))
print(a)
print(b)
```

```python
val = signal.convolve(a,b[::-1])
val
```

```python
val = val[1:] - val[:-1]
val
```

```python
jump = []
if abs(max(val)) > abs(min(val)):
    jump =  np.where(val == max(val))
    print(max(val) , jump)
else:
    jump =  np.where(val == min(val))
    print(min(val) , jump)
```

```python
for j in jump:
    if j < len(a) - 1:
        k = j[0] + 1
        
        print(np.dot(a[:k],b[-k:]))
        print(a[:k])
        print(b[-k:])
        
        print('\n')
        print(np.dot(a[:k+1],b[-k-1:]))
        print(a[:k+1])
        print(b[-k-1:])
        
    if j[0] > len(a) - 1:
        k = len(a)*2 - 1 - j[0]
        
        print(np.dot(a[-k:],b[:k]))
        print(a[-k:])
        print(b[:k])
        
        print('\n')
        print(np.dot(a[-k+1:],b[:k-1]))
        print(a[-k+1:])
        print(b[:k-1])
```
