```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import time
```

```python
# Necessary declarations
val_map = {
    0 : [0,0,0,1],
    1 : [0,0,1,0],
    2 : [0,1,0,0],
    3 : [1,0,0,0]
}
size = 200
top_jumps = 50
top_vals = 50
```

```python

######################### Made two sequences and calculated the convolution

# a = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
# b = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
# res = []
# for i in range(size):
#     res += val_map[a[i]]
# a_h = res
# res = []
# for i in range(size):
#     res += val_map[b[i]]
# b_h = res
# res = signal.convolve(a_h,b_h[::-1])[3::4]
# print(res)
```

```python
x_points = []
y_points = []
y_points2 = []
start = time.time()
for k in range(100):
    
    ######################### Made two sequences and calculated the convolution

    a = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
    b = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
    res = []
    for i in range(size):
        res += val_map[a[i]]
    a_h = res
    res = []
    for i in range(size):
        res += val_map[b[i]]
    b_h = res
    res = signal.convolve(a_h,b_h[::-1])[3::4]
    # print(res)
    
    ######################### Finding the big jumps.
    
    diff = []
    for i in range(1,len(res)):
        diff.append(abs(res[i] - res[i-1]))
    diff = np.array(diff)
    diff_sort = np.sort(diff)

    peaks = []
    for i in range(1,top_jumps):
        peaks += list(np.where(diff==diff_sort[-i])[0])
    # print(len(peaks))
    # print(diff)
    
    ######################### Calculating the largest cluster length at big jumps.
    c_length = 0
    for x in peaks:
        if res[x+1] > res[x]:
            i = x+2
            # print("Second one")
        else:
            i = x+1
            # print("First one")
        
        truth_val = []
        if i <= len(a):
            truth_val = a[:i] == b[-i:]
        else:
            i = i - len(a)
            truth_val = a[i:] == b[:-i]
        
        temp = 0
        for i in range(len(truth_val)):
            if truth_val[i]:
                temp += 1
            else:
                temp = 0
            c_length = max(c_length,temp)

    ######################### Finding the cluster length at the highest values of the convolution

    res_sort = np.sort(res)
    high = []
    for i in range(1,top_vals):
        high.append(np.where(res==res_sort[-i])[0][0])

    h_length = 0
    for x in high:
        if res[x+1] > res[x]:
            i = x+2
        else:
            i = x+1
        
        truth_val = []
        if i <= len(a):
            truth_val = a[:i] == b[-i:]
        else:
            i = i - len(a)
            truth_val = a[i:] == b[:-i]

        temp = 0
        for i in range(len(truth_val)):
            if truth_val[i]:
                temp += 1
            else:
                temp = 0
            h_length = max(h_length,temp)

    ######################### Finding the largest cluster of the sequence as a whole
    length = 0
    clusters = []
    for s in range(1,len(a)):
        flag1 = True
        for i in range(len(a)+1-s):
            flag2 = False
            for j in range(len(b)+1-s):
                arr = a[i:i+s]==b[j:j+s]
                if arr.all():
                    clusters.append(a[i:i+s])
                    length = s
                    flag1 = False
                    flag2 = True
                    break
            if flag2:
                break
        if flag1:
            break
    
    
    # print(clusters)
    # print(length,c_length,h_length)
    x_points.append(length)
    y_points.append(c_length)
    y_points2.append(max(c_length, h_length))
    # print(k)
    
end = time.time()
print(end-start)
grad = np.dot(y_points,x_points)/np.dot(x_points,x_points)

print(grad)
```

```python
l = 15
plt.scatter(x_points,y_points2,alpha = 0.2)
plt.plot([i for i in range(l)],[i for i in range(l)])
plt.plot([i for i in range(l)],[i*grad for i in range(l)])
# plt.savefig("200 elements,m=0.89.png")
```
