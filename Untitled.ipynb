```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
val_map = {
    0 : [0,0,0,1],
    1 : [0,0,1,0],
    2 : [0,1,0,0],
    3 : [1,0,0,0]
}
size = 100
```

```python
a = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
b = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
res = []
for i in range(size):
    res += val_map[a[i]]
a_h = res
res = []
for i in range(size):
    res += val_map[b[i]]
b_h = res
print(a,b)
print(a_h,b_h)
```

```python
print(a)
print(b)
res = signal.convolve(a_h,b_h[::-1])[3::4]
print(res)
```

```python
diff = []
for i in range(1,len(res)):
    diff.append(abs(res[i] - res[i-1]))
diff = np.array(diff)

peaks = np.where(diff==max(diff))
print(max(diff))
for x in peaks[0]:
    print("\n")
    i = x+1
    if i < len(a):
        print(a[:i])
        print(b[-i:])
    else:
        i = i - len(a)
        print(a[i:])
        print(b[:-i])

    i = x+2
    if i < len(a):
        print(a[:i])
        print(b[-i:])
    else:
        i = i - len(a)
        print(a[i:])
        print(b[:-i])
```
