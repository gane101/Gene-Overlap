```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import time
```

```python
# Necessary declarations
val_map = {
    0 : [0,0,0,1],
    1 : [0,0,1,0],
    2 : [0,1,0,0],
    3 : [1,0,0,0]
}
size = 155
```

```python
x_points = []
y_points = []
start = time.time()
for k in range(100):

    ######################### Made two sequences and calculated the convolution
    
    a = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
    b = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
    res = []
    for i in range(size):
        res += val_map[a[i]]
    a_h = res
    res = []
    for i in range(size):
        res += val_map[b[i]]
    b_h = res
    res = signal.convolve(a_h,b_h[::-1])[3::4]
    # print(res)
    
    ######################### Finding the big jumps.
    
    diff = []
    for i in range(1,len(res)):
        diff.append(abs(res[i] - res[i-1]))
    diff = np.array(diff)
    
    peaks = np.where(diff==max(diff))
    # print(diff)
    
    ######################### Calculating the largest cluster length at big jumps.
    for x in peaks[0]:
        if res[x+1] > res[x]:
            i = x+2
            # print("Second one")
        else:
            i = x+1
            # print("First one")
        
        truth_val = []
        if i <= len(a):
            truth_val = a[:i] == b[-i:]
        else:
            i = i - len(a)
            truth_val = a[i:] == b[:-i]
        
        temp = 0
        c_length = 0
        for i in range(len(truth_val)):
            if truth_val[i]:
                temp += 1
            else:
                temp = 0
            c_length = max(c_length,temp)
    
    ######################### Finding the largest cluster of the sequence
    length = 0
    clusters = []
    for s in range(1,len(a)):
        flag1 = True
        for i in range(len(a)+1-s):
            flag2 = False
            for j in range(len(b)+1-s):
                arr = a[i:i+s]==b[j:j+s]
                if arr.all():
                    clusters.append(a[i:i+s])
                    length = s
                    flag1 = False
                    flag2 = True
                    break
            if flag2:
                break
        if flag1:
            break

    
    # print(clusters)
    # print(length,c_length)
    x_points.append(length)
    y_points.append(c_length)
    # print(k)
    
end = time.time()
print(end-start)
reciprocal = np.vectorize(lambda x:1/x)
rec = reciprocal(x_points)
grad = np.dot(y_points,rec)/len(x_points)
print(grad)
```

```python
l = 15
plt.scatter(x_points,y_points,alpha = 0.2)
plt.plot([i for i in range(l)],[i for i in range(l)])
plt.plot([i for i in range(l)],[i*grad for i in range(l)])
```
