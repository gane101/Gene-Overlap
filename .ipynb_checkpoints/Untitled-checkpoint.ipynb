```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
val_map = {
    0 : [0,1],
    1 : [1,0]
}
size = 10
```

```python
a = np.array([int(np.floor(np.random.random()*2)) for i in range(size)])
b = np.array([int(np.floor(np.random.random()*2)) for i in range(size)])
res = []
for i in range(size):
    res += val_map[a[i]]
a_h = res
res = []
for i in range(size):
    res += val_map[b[i]]
b_h = res
print(a,b)
print(a_h,b_h)
```

```python
print(a)
print(b)
res = signal.convolve(a_h,b_h[::-1])[1::2]
print(res)
```

```python
diff = []
diff
for i in range(1,len(res)):
    diff.append(abs(res[i] - res[i-1]))
diff = np.array(diff)
print(res)
print(diff)
peaks = np.where(diff==max(diff)) + np.where(diff==min(diff))
peaks
```
