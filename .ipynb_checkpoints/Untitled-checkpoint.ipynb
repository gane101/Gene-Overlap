```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Necessary declarations
val_map = {
    0 : [0,0,0,1],
    1 : [0,0,1,0],
    2 : [0,1,0,0],
    3 : [1,0,0,0]
}
size = 20
```

```python
# Made two sequences and calculated the convolution
a = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
b = np.array([int(np.floor(np.random.random()*4)) for i in range(size)])
res = []
for i in range(size):
    res += val_map[a[i]]
a_h = res
res = []
for i in range(size):
    res += val_map[b[i]]
b_h = res
res = signal.convolve(a_h,b_h[::-1])[3::4]
print(res)
```

```python
# Finding the big jumps and looking at the actual jumps
diff = []
for i in range(1,len(res)):
    diff.append(abs(res[i] - res[i-1]))
diff = np.array(diff)

peaks = np.where(diff==max(diff))
print(max(diff))
for x in peaks[0]:
    print("\n")
    i = x+1
    if i <= len(a):
        print(a[:i])
        print(b[-i:])
    else:
        i = i - len(a)
        print(a[i:])
        print(b[:-i])
    
    i = x+2
    if i < len(a):
        print(a[:i])
        print(b[-i:])
    else:
        i = i - len(a)
        print(a[i:])
        print(b[:-i])
a,b
```

```python
x = peaks[0][0] + 1
print(x)
```

```python
# Finding the largest cluster of the sequence
length = 0
clusters = []
for s in range(1,len(a)):
    for i in range(len(a)+1-s):
        for j in range(len(b)+1-s):
            # print(a[i:i+s]==b[j:j+s] , a[i:i+s] , b[j:j+s])
            arr = a[i:i+s]==b[j:j+s]
            if arr.all():
                clusters.append(a[i:i+s])
                if length != s:
                    length = s
                    clusters = [a[i:i+s]]
print(length)
print(clusters)
```
